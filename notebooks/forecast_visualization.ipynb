```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../outputs/forecast_results.csv")

plt.figure(figsize=(14,5))
plt.plot(df["actual_price"], label="Actual Price")
plt.plot(df["predicted_price"], label="Predicted Price")

plt.title("Actual vs Predicted Day-Ahead Power Prices")
plt.xlabel("Time")
plt.ylabel("€/MWh")
plt.legend()

plt.show()
```
